```python
def anagram(str1, str2):
    str1_obj = {}
    str2_obj = {}
    str1 = str1.lower()
    str2 = str2.lower()
    for l in str1:
        if l in str1_obj:
            str1_obj[l] += 1
        else:
            str1_obj[l] = 1
    for l in str2:
        if l in str2_obj:
            str2_obj[l] += 1
        else:
            str2_obj[l] = 1
    if str1_obj == str2_obj:
        return True
    else:
        return False
```

```python
anagram("my name is l XLAX", "JBSJ KS")
```

```python
anagram("KARTIK", "karitik")
```

```python
def pairs(arr, no):
    seen = set()
    output = set()
    for num in arr:
        target = no - num
        if target not in seen:
            seen.add(num)
        else:
            output.add((target, num))
    return output
```

```python
pairs([1,3,2,2], 4)
```

```python
def profit(stock_prices):
    min_stock = stock_prices[0]
    max_profit = 0
    
    for price in stock_prices:
        min_stock = min(min_stock, price)
        
        comp_profit = price - min_stock
        
        max_profit = max(max_profit, comp_profit)
        
    return max_profit
        
    
```

```python
profit([12,11,15,3,10])
```
