### Given a sorted array. Find three values in the array that sum to a given number. E.g. given [1,2,3,4,5], can you find three values that sum to 10? If so, what values are they? The solution has to be better than O(n^3)  1 Answer


```python
arr = [1,2,3,4,5]
def find3(arr, sum):
    for i in range(0, len(arr)-1):
        first = i + 1
        last = len(arr) - 1
        while(first < last):
            if(arr[i] + arr[first] + arr[last] == sum):
                return arr[i], arr[first], arr[last]
            elif(arr[i] + arr[first] + arr[last] > sum):
                last-=1
            elif(arr[i] + arr[first] + arr[last] < sum):
                first+=1
            
    return False
```

### Given an array of numbers, can you find the indexes of a subarray, that if you sorted the subarray, would result in the entire array being sorted?

### e.g. Given [1,2,3,8,7,6,5,4,9] Return 4 and 8  

```python
Public class Solution {
    public int findUnsortedSubarray(int[] nums) {
        int[] snums = nums.clone();
        Arrays.sort(snums);
        int start = snums.length, end = 0;
        for (int i = 0; i < snums.length; i++) {
            if (snums[i] != nums[i]) {
                start = Math.min(start, i);
                end = Math.max(end, i);
            }
        }
        return (end - start >= 0 ? end - start + 1 : 0);
    }
}
```

```python
def p_soln(arr):
    
    duplicate = list(arr)
    duplicate.sort()
    
    start = end = 0
    for i in range(len(arr)):
        if arr[i] != duplicate[i]:
            start = i
            break
            
    for i in range(len(arr)-1,0,-1):
        if arr[i] != duplicate[i]:
            end = i
            break
            
    return (start,end)
```

```python
def dist(p0, p1):
    return (((p0[0] - p1[0])**2) + ((p0[1] - p1[1])**2))
def shortestDist(points):
        sh = float("inf")
        for i in range(1, len(points)):
                d = dist(points[i-1], points[i])
                if d < sh:
                        sh = d
        return sh
```

```python
from collections import defaultdict
def solution1(A):
    totalElements = len(set(A))
    start = end = 0
    elements = defaultdict(int)
    minDays = 999999
    while len(elements) < totalElements and end < len(A):
        elements[ A[end] ] += 1
        end += 1

        #while len(elements) >= totalElements:
        for _ in range(0,totalElements):
            if len(elements) >= totalElements:
                minDays = min( minDays, end-start )
                removeElem = A[start]
                elements[ removeElem ] -= 1
                start += 1
                if 0 == elements[ removeElem ]:
                    elements.pop( removeElem )
                    break
    return minDays
```
